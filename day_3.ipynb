```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3]
y = [3, 6, 1]
plt.plot(x, y)
plt.title("Sample Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = np.arange(0, 2*(np.pi), 0.01)
y = np.sin(x)
plt.plot(x, y)
plt.plot(x, np.cos(x))
plt.plot(x, x**2)
plt.title("Sin graph")
plt.xlabel("X")
plt.ylabel("Sin(X)")
plt.show()
```

```python
x = [10, 12, 31, 24, 5]
y = ['Dhaka', 'Chattogram', 'Sylhet', 'Khulna', 'Barishal']
plt.bar(y, x)
plt.savefig("sample01.png")
plt.show()
```

```python
fig = plt.figure()
plt1 = fig.add_subplot(221)
plt2 = fig.add_subplot(222)
plt3 = fig.add_subplot(223)
plt4 = fig.add_subplot(224)

x = np.arange(-10, 10, 0.01)
plt1.plot(x, x)
plt1.set_title('$y_1 = x$')

plt2.plot(x, x**2)
plt2.set_title('$y_2 = x^2$')

plt3.plot(x, x**3)
plt3.set_title('$y_3 = x^3$')

plt4.plot(x, x**4)
plt4.set_title('$y_4 = x^4$')

fig.subplots_adjust(hspace=0.5,wspace=0.5)

plt.show()
```
